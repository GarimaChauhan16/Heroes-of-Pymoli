### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Total number of players is the number of unique player names
Total_Player_Count = purchase_data['SN'].nunique()

pd.DataFrame({"Total Players": [Total_Player_Count]})
```

```python
# Some of the same items names have two different item IDs, so the number of unique Item names is not same as Number of unique IDs

# Combining Item name and IDs gives a new column 'Item name & ID' and can be used to find out unique items
purchase_data['Item Name & ID'] = purchase_data['Item ID'].map(str)  + "- " + purchase_data['Item Name']

# Ptint Table
purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Number of unique items is the number of unique 'Item name & ID' 
Item_Count = purchase_data['Item Name & ID'].nunique()

Avg_Price = purchase_data['Price'].mean()

Total_Purchase = len(purchase_data)

# Sum of total price is the total revenue
Total_Revenue = purchase_data['Price'].sum()

# Create a summary data frame
Summary_Table = pd.DataFrame({"Number of Unique Items": [Item_Count], "Average Price": [Avg_Price], "Number of Purchases": [Total_Purchase], "Total Revenue": [Total_Revenue]})

# Data formatting
Summary_Table["Average Price"] = Summary_Table["Average Price"].map("$ {:,.2f}".format)

Summary_Table["Total Revenue"] = Summary_Table["Total Revenue"].map("$ {:,}".format)

# Ptint Table
Summary_Table
```

# Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Grouping the items by gender
grouped_purchase_data = purchase_data.groupby(['Gender'])

# Unique number of players grouped by gender gives the player count for each gender
player_Count_by_gender = grouped_purchase_data.nunique()["SN"]

percent_player_Count = (player_Count_by_gender/Total_Player_Count) * 100

# Create a data frame to show the Gender Demographics 
Gender_Demographics = pd.DataFrame({"Total Count": player_Count_by_gender, "Percentage of Players": percent_player_Count})

# Sort the data frame by descending number of player count
Gender_Demographics = Gender_Demographics.sort_values("Total Count", ascending=False)

# Data formatting
Gender_Demographics["Percentage of Players"] = Gender_Demographics["Percentage of Players"].map("{:.2f}".format)

# Ptint Table
Gender_Demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Grouping the items by gender for gender based Purchasing analysis
grouped_purchase_data = purchase_data.groupby(['Gender'])

# Purchase count for each gender
purchase_count_by_gender = grouped_purchase_data.count()['Purchase ID']

# Total Purchase Value for each gender is the sum of purchase price
total_purchase_value_by_gender = grouped_purchase_data.sum()['Price']

# Average Purchase Price for each gender is the total purchase value divided by purchase count
avg_price_by_gender = total_purchase_value_by_gender/ purchase_count_by_gender

# Avg Total Purchase per Person for each gender is total purchase value divided by the player count
avg_purchase_per_person_by_gender = total_purchase_value_by_gender/ player_Count_by_gender

# Create a data frame for gender based purchasing analysis 
Purchasing_Analysis_Gender = pd.DataFrame({"Purchase Count" : purchase_count_by_gender, "Average Purchase Price": avg_price_by_gender, 
                                    "Total Purchase Value": total_purchase_value_by_gender,
                                    "Avg Total Purchase per Person": avg_purchase_per_person_by_gender})

# Data formatting
Purchasing_Analysis_Gender["Average Purchase Price"] = Purchasing_Analysis_Gender["Average Purchase Price"].map("$ {:.2f}".format)
Purchasing_Analysis_Gender["Total Purchase Value"] = Purchasing_Analysis_Gender["Total Purchase Value"].map("$ {:,.2f}".format)
Purchasing_Analysis_Gender["Avg Total Purchase per Person"] = Purchasing_Analysis_Gender["Avg Total Purchase per Person"].map("$ {:.2f}".format)

# Ptint Table
Purchasing_Analysis_Gender
```

## Age Demographics

* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
group = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], bins, labels=group)
purchase_data.head()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Grouping the items by names
Name_group_purchase_data = purchase_data.groupby(['SN'])

# Purchase count for each player
purchase_count_by_name = Name_group_purchase_data.count()['Purchase ID']

# Total Purchase Value for each player is the sum of purchase price
total_purchase_value_by_name = Name_group_purchase_data.sum()['Price']

# Average Purchase Price for each player is the total purchase value divided by purchase count
avg_price_by_name = total_purchase_value_by_name/ purchase_count_by_name

# Create a data frame for purchasing analysis of each player
Purchasing_Analysis_Name = pd.DataFrame({"Purchase Count" : purchase_count_by_name, "Average Purchase Price": avg_price_by_name, 
                                    "Total Purchase Value": total_purchase_value_by_name})

# Sorting the data by total purchase values gives the information about top spenders 
top_spenders = Purchasing_Analysis_Name.sort_values("Total Purchase Value", ascending=False)

# Data formatting
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("$ {:.2f}".format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("$ {:,.2f}".format)

# Ptint Table
top_spenders.head()                 
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Grouping the items by Item Id and Item names
Item_group_purchase_data = purchase_data.groupby(['Item ID', 'Item Name'])

# Purchase count for each item
purchase_count_by_item = Item_group_purchase_data.count()['Purchase ID']


item_price = Item_group_purchase_data.mean()['Price']

# Total Purchase Value for each item is the sum of purchase price
total_purchase_value_by_item = Item_group_purchase_data.sum()['Price']

# Create a data frame for purchasing analysis of each item
Purchasing_Analysis_Item = pd.DataFrame({"Purchase Count" : purchase_count_by_item, "Item Price": item_price, 
                                    "Total Purchase Value": total_purchase_value_by_item})

# Sorting the data by purchase count gives the information about the most popular item
Most_Popular_Items = Purchasing_Analysis_Item.sort_values("Purchase Count", ascending=False)

# Data formatting
Most_Popular_Items["Item Price"] = Most_Popular_Items["Item Price"].map("$ {:.2f}".format)
Most_Popular_Items["Total Purchase Value"] = Most_Popular_Items["Total Purchase Value"].map("$ {:,.2f}".format)

# Ptint Table
Most_Popular_Items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sorting the data by total purchase value gives the information about the most profitable item
Most_Profitable_Items = Purchasing_Analysis_Item.sort_values("Total Purchase Value", ascending=False)

# Data formatting
Most_Profitable_Items["Item Price"] = Most_Profitable_Items["Item Price"].map("$ {:.2f}".format)
Most_Profitable_Items["Total Purchase Value"] = Most_Profitable_Items["Total Purchase Value"].map("$ {:,.2f}".format)

# Ptint Table
Most_Profitable_Items.head()
```
